### Sampling Distribution
---

$(X_1,X_2,..)$ are independent random variables with uniform distribution $X_i \sim$ **Unif(0,1)**. We define the statistic as $\mathbb{\bar{X}_n}=\frac{1}{n}\sum^{n}_{i=1}X_i$.

The Uniform distribution is given as :
$$
 f(x) = \left\{\begin{array}{lr}
        \frac{1}{b-a}, & \text{for } x \in [a,b]\\
        0, &  otherwise\\
        \end{array}\right\}
        $$
Expectation for uniform distribution is:

$$
    \begin{align}
        \mathbb{E}{(X)} &= \int^b_a xf(x)\mathrm{d}x\\
        &=\frac{1}{b-a}\left[\frac{x^2}{2}\right]^b_a\\
        &=\frac{1}{2}(b+a)
    \end{align}
$$



$$
    \begin{align}
        \mathbb{E}{(X^2)} &= \int^b_a x^2f(x)\mathrm{d}x\\
        &=\frac{1}{b-a}\left[\frac{x^3}{3}\right]^b_a\\
        &=\frac{1}{3(b-a)}(b^3-a^3)\\
        &=\frac{1}{3}(a^2+b^2+ab)
    \end{align}
$$

Variance of Uniform distribution is:

$$ 
    \begin{align}
        \mathbb{V}{(X)} &= \mathbb{E}(X^2) - (\mathbb{E}{(X)})^2\\
        &= \frac{1}{3}(a^2+b^2+ab) - \left(\frac{1}{2}(b+a)\right)^2\\
        &= \frac{4b^2+4a^2+4ab - 3a^2-3b^2-6ab}{12}\\
        &=\frac{1}{12}(b-a)^2
    \end{align}
$$



For Unif(0,1):

$$ 
    \mathbb{E}{(X)} = 0.5 = \mu\\
    \mathbb{V}{(X)} = \frac{1}{12} = 0.08 = \sigma^2
$$

Now, we know sample mean and variance of sample means is given by:

$$ 
   \mathbb{E}{(\mathbb{\bar{X}_n})} =  \mu\\ 
   \mathbb{V}{(\mathbb{\bar{X}_n})} = \frac{\sigma^2}{n}
$$




```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
'''Simulating Expectation of sample mean and variance of sample mean as a function of n'''

n=1000
x=0
var_temp=0
x_bar = []
x_var = []
y = np.random.uniform(0,1,n+1)
y.size
for i in range(n+1):
    x+=y[i]
    x_mean = x/(i+1)
    x_bar.append(x_mean)
mean = sum(x_bar)/len(x_bar)
for s in x_bar:
    var_temp=(s-mean)**2
    x_var.append(var_temp)
    
plt.xlabel('n')
plt.ylabel('Sample Mean')
plt.plot(list(range(n+1)), x_bar, label='Sample mean vs n')
plt.legend()
plt.show()
plt.xlabel('n')
plt.ylabel('Variance of Sample Mean')
plt.plot(list(range(n+1)),x_var,label='Variance of sample mean vs n ')
plt.legend()
```

We observe from above graphs as n increases $\mathbb{E}{(\mathbb{\bar{X}_n})}$ approaches $\mu$ according to law of large numbers and $\mathbb{V}{(\mathbb{\bar{X}_n})}$ decreases as n increases.

#### Sampling distribution of $\mathbb{\bar{X}_n}$

```python
sample1 = [(np.sum(np.random.uniform(0,1,1)))/1 for _ in range(500)]
sample5 = [(np.sum(np.random.uniform(0,1,5)))/5 for _ in range(500)]
sample25 = [(np.sum(np.random.uniform(0,1,25)))/25 for _ in range(500)]
sample100 = [(np.sum(np.random.uniform(0,1,100)))/100 for _ in range(500)]

fig, axs = plt.subplots(2, 2)
axs[0,0].hist(sample1,bins=10)
axs[0,0].set_title('n=1')
axs[0,1].hist(sample5,bins=10)
axs[0,1].set_title('n=5')
axs[1,0].hist(sample25,bins=10)
axs[1,0].set_title('n=25')
axs[1,1].hist(sample100,bins=10)
axs[1,1].set_title('n=100')
axs[0,0].set_xlabel('mean')
axs[0,1].set_xlabel('mean')
axs[1,1].set_xlabel('mean')
axs[1,0].set_xlabel('mean')
plt.tight_layout()
plt.show()
```

We can see that as n increases sampling distribution of the statistic(mean) becomes normally distributed and variance decreases as the mean becomes more centered as per calculations.
