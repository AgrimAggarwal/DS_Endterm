## Question 4
### Central Limit Theorem
---

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Here, we verify the central limit theorem for different distributions. Given enough samples, it does not matter from which distribution the random variables are taken from as long as they are independent and identically distributed, the mean of those random variables will approach normal distribution as the size of sample increases.

#### For Uniform Distribution : 

```python
'''x is a vector of size 3000, with random variables generated from uniform distribution'''

x= np.random.uniform(0,1,3000)  

samplesx1 = [np.mean(np.random.choice(np.hstack((x)),size=1)) for _ in range(150)] #sample size 1
samplesx2 = [np.mean(np.random.choice(np.hstack((x)),size=10)) for _ in range(150)] #sample size 10
samplesx3 = [np.mean(np.random.choice(np.hstack((x)),size=50)) for _ in range(150)] #sample size 50
samplesx4 = [np.mean(np.random.choice(np.hstack((x)),size=200)) for _ in range(150)] #sample size 200 (all taken 150 times)

fig,axs=plt.subplots(2,2)
axs[0,0].hist(samplesx1,bins=10)
axs[0,0].set_title('n=20 ')
axs[0,1].hist(samplesx2,bins=10)
axs[0,1].set_title('n=10')
axs[1,0].hist(samplesx3,bins=10)
axs[1,0].set_title('n=50')
axs[1,1].hist(samplesx4,bins=10)
axs[1,1].set_title('n=200')
plt.tight_layout()
plt.show()
```

<u> We can clearly see the sampling distribution of the mean of random variables taken from uniform distribution approaches uniform distribution as sample size increases. The mean also becomes more centered (convergence in mean from law of large numbers) and variance of sample mean also decreases as $\mathbb{V}{(\mathcal{\bar{X}_n})} = \frac{\sigma^2}{n}$

#### For Poisson Distribution

```python
y=np.random.poisson(2,3000)
samplesy1 = [np.mean(np.random.choice(np.hstack((y)),size=1)) for _ in range(150)] #sample size 1
samplesy2 = [np.mean(np.random.choice(np.hstack((y)),size=10)) for _ in range(150)] #sample size 10
samplesy3 = [np.mean(np.random.choice(np.hstack((y)),size=50)) for _ in range(150)] #sample size 50
samplesy4 = [np.mean(np.random.choice(np.hstack((y)),size=200)) for _ in range(150)] #sample size 200 (all taken 150 times)

fig,axs=plt.subplots(2,2)
axs[0,0].hist(samplesy1,bins=10)
axs[0,0].set_title('n=20 ')
axs[0,1].hist(samplesy2,bins=10)
axs[0,1].set_title('n=10')
axs[1,0].hist(samplesy3,bins=10)
axs[1,0].set_title('n=50')
axs[1,1].hist(samplesy4,bins=10)
axs[1,1].set_title('n=200')
plt.tight_layout()
plt.show()
```

#### For Exponential Distribution

```python
z=np.random.poisson(2,3000)
samplesz1 = [np.mean(np.random.choice(np.hstack((z)),size=1)) for _ in range(150)] #sample size 1
samplesz2 = [np.mean(np.random.choice(np.hstack((z)),size=10)) for _ in range(150)] #sample size 10
samplesz3 = [np.mean(np.random.choice(np.hstack((z)),size=50)) for _ in range(150)] #sample size 50
samplesz4 = [np.mean(np.random.choice(np.hstack((z)),size=200)) for _ in range(150)] #sample size 200 (all taken 150 times)

fig,axs=plt.subplots(2,2)
axs[0,0].hist(samplesz1,bins=10)
axs[0,0].set_title('n=20 ')
axs[0,1].hist(samplesz2,bins=10)
axs[0,1].set_title('n=10')
axs[1,0].hist(samplesz3,bins=10)
axs[1,0].set_title('n=50')
axs[1,1].hist(samplesz4,bins=10)
axs[1,1].set_title('n=200')
plt.tight_layout()
plt.show()
```

Above was the example of classic CLT where all the random variables were independent and identically distributed. The identically distributed condition can suppressed by the condition on CLT provided by Lyapunov called Lyapunov Central Limit Theorem.

We demonstrate this by taking random variables from different distributions in a fixed interval and taking their mean and as we increase the sample size, we see that it approaches normal distribution.

#### Taking a sample containing of uniform, poisson and exponential distribution
