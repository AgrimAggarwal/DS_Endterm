### Page Rank Algorithm
---

Page Rank is the algorithm published by founders of Google in order to estimate the heirarchy of web pages on the search engine. It essentially ranks web pages on the basis of how important/referenced that particular website is. It can be understood as a Markov Process where each webpage is a state and the links between webpages can be thought of as transitions.

Here, we have implemented the algorithm using the following iterative formula:

$$ 
    PR(p_i) = \frac{1-d}{N} + d\cdot\left(\sum_{p_j\in in(p_i)}\frac{PR(p_j)}{|out(p_j)|}\right)
$$

> $p_i$ are the pages/states/nodes under consideration\
>$in(p_j)$ are the pages which points towards(inlink pages) $p_i$.\
>$|out(p_j)|$ is the number of outbound pages from $p_j$.\
>$N$ is the total number of pages.\
>$d$ is the damping factor.\
>$PR(p_j)$ on the right hand side is the page rank of that page from previous iteration.\

$d$ is introduced so that pages with no incoming links do not have 0 page rank and the total probability falls short. It also ensures, that the surfer is not stuck at part of a graph.

In this implementation, to handle sinks, 'adding edges method' has been used. Sinks are nodes that do not have any outgoing links. These do not provide any rank to other pages and this results in accumulation if we consider the analogy of page rank being like a fluid. We solve this issue by adding edges from the sink to every other page, so as to distribute rank evenly among other pages.

After implementation of the self-written algorithm, we also test it using the built-in library to calculate page rank.


```python
import networkx as nx    #Python library to create directed graphs
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
'''add_edges is the function to randomly add edges to nodes'''

def add_edges(G,p):
    for i in G.nodes():
        for j in G.nodes():
            if i!=j:
                r=np.random.random()
                if r<=p:
                    G.add_edge(i,j)
                else:
                    continue

'''handle_sinks is the function to add edges from sinks to every other node '''                    
def handle_sinks(G):
    for i in G.nodes():             #loop throgh every node of the graph
        if len(G.out_edges(i))==0:  #condition for if a node does not have any outgoing link: identifying sink
            for j in G.nodes():
                if i!=j:
                    G.add_edge(i,j) #connect the sink to every other node
                else:
                    continue
        else:
            continue

'''start_vector is the function to initialize the ranks evenly among all pages'''
def start_vector(G):
    number = len(G.nodes())                        #no. of nodes
    ranks=[(1/number) for _ in range(number)]      #creating the initial rank vector 
    return ranks
    
'''calc_rank is the function to calculate rank with damping factor=0.85 as default'''           
def calc_rank(G,ranks,d=0.85):
    previous_ranks= ranks                          #takes the rank vector of previous iteration
    rank= (1-d)/(len(G.nodes()))                   #first part of iterative formula 
    new_ranks=[]                                   #empty rank vector for current iteration
    
    
    for i in G.nodes():                            #looping through each node
        in_rank_temp=0
        inlinks = G.in_edges(i)                    #identifying the incoming links to the node in consideration 
        for ik in inlinks:                         # looping through each inlink node
            outlink = len(G.out_edges(ik[0]))      #no. of outgoing nodes from incoming node to node in consideration
            in_rank_temp+=(previous_ranks[ik[0]]/outlink)  #second part of iterative formula
        rank1=rank+d*(in_rank_temp)   
        new_ranks.append(rank1)                    #appending rank for this iteration to new_ranks list
    
            
    return G,new_ranks

'''convergence_ranks is the function to iterate over calc_rank until convergence(here we take it to be 20 iterations
since it usually converges in that many iterations)'''

def convergence_ranks(G,ranks):
    previous_ranks = ranks
    iterations=0
    while(iterations<20):
        G,new_ranks = calc_rank(G,previous_ranks,d=0.85)
        previous_ranks=new_ranks
        print('Iteration no.:', iterations+1)
        print(previous_ranks)
        print("")
        iterations+=1
    
    return G,new_ranks
        
    

```

#### Here we create a directed graph without handle_sinks. Since graphs are created with edges added randomly, sometimes a sink may appear and we correct that in the next cell. We also create the adjacency matrix for this graph which shows connections between nodes with a (0,1) graph. Rows being outgoing and columns being inlink.

Note:- We can run this cell till a graph appears which has sinks.

```python
G= nx.DiGraph()      #using networkx library to create the directed graph
G.add_nodes_from([i for i in range(10)])   #creating nodes. (number can be changed as per liking)
add_edges(G,0.3)                           #adding edges with a probability of 0.3. can be changed
nx.draw(G,with_labels=True)           
plt.show()
A = nx.adjacency_matrix(G)
print('Adjacency matrix for this graph is in the order (0,1,2,3,..,9)')
print('')
print(A.todense())
```

#### Here we use the handle_sinks function to add outgoing edges for every sink (if any).

```python
handle_sinks(G)

nx.draw(G,with_labels=True)
plt.show()
A = nx.adjacency_matrix(G)
print('Adjacency matrix for this graph in the order (0,1,2,...,9)')
print('')
print(A.todense())
```

#### Now we implement our page rank algorithm

```python
ranks=start_vector(G)    #initializing the initial ranks vector. distributed evenly
G,ranks = convergence_ranks(G,ranks)  #run the algorithm till convergence

'''built-in pagerank library to check the validity of our solution(assuming built-in is correct)'''
pr = nx.pagerank_scipy(G, alpha=0.85,tol=1e-6)
print('Page Rank using built-in library')
print('')
print(pr)
```
